```python
# Carica la API key dai Colab Secrets
#from google.colab import userdata     #per usarlo su colab
import anthropic
import os
from dotenv import load_dotenv  #per usarlo in locale 

# Legge il secret che hai configurato
# os.environ["ANTHROPIC_API_KEY"] = userdata.get("ANTHROPIC_API_KEY")    #per usarlo su colab

# Carica le variabili d'ambiente dal file .env
load_dotenv()                                                #per usarlo in locale 

# Crea il client                                                                <----------
client = anthropic.Anthropic()

# Verifica (senza stampare la chiave!)
key = os.environ.get("ANTHROPIC_API_KEY", "")
if key.startswith("sk-ant"):
    print(f"✅ API key configurata correttamente: sk-ant-...{key[-6:]}")
else:
    print("❌ API key non trovata. Controlla i Colab Secrets.")
```

```python
def chiedi_claude(domanda, temperature=0.7, system=None, max_tokens=300):
    params = {
        "model": "claude-haiku-4-5-20251001",
        "max_tokens": max_tokens,
        "temperature": temperature,
        "messages": [{"role": "user", "content": domanda}]
    }
    if system:
        params["system"] = system

    response = client.messages.create(**params)

# Inserisci il return corretto                                                  <----------
    return response.content[0].text
```

```python
# Chiedere all'utente la domanda o curiosità   <----------

domanda = input("Inserisci la tua domanda o curiosità: ")
```

```python
# Chiedi all'utente se deve rispondere a un bambino o a un esperto              <----------

ruolo = input("Deve rispondere a un bambino o a un esperto?: ")

if ruolo == "bambino":
    system = "Sei un insegnante per bambini di 3 anni."

elif ruolo == "esperto":
    system = "Sei un professore universitario."
```

```python
# Inserisci i parametri da passare a chiedi_claude()                            <----------
risposta = chiedi_claude(domanda, system=system)
print (risposta)
```
